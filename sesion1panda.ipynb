```python
#Quiero crear una nueva columna que se llama "Categoria" y que tenga el valor de "Adulto" si la edad es entre 18 y 59 años, "Niño" si la edad es entre 0 y 16 año, "Adolecente" si la edad es entre 12 y 17 años, "Adulto Mayo" si la edad es mayor de 60 años
df_titanic2["Categoria"] = "Adulto"
df_titanic2.loc[(df_titanic2["Age"] < 12) & (df_titanic2["Age"] >0), "Categoria"] = "Niño"
df_titanic2.loc[(df_titanic2["Age"] < 18) & (df_titanic2["Age"] >=12), "Categoria"] = "Adolecente"
df_titanic2.loc[(df_titanic2["Age"] >= 60), "Categoria"] = "Adulto Mayo"
df_titanic2["Categoria"].value_counts()
```

```python
#Crear una nueva columuna que se llama "Edad" y que tenga el valor de la edad en meses
df_titanic2["Edad"] = df_titanic2["Age"]
```

```python
#Crear una funcion que me diga si la edad es menor de 12 años es "Niño", si la edad es mayor de 12 años y menor de 18 años es "Adolecente", si la edad es mayor de 18 años y menor de 60 años es "Adulto", si la edad es mayor de 60 años es "Adulto Mayo"
def edad_categoria(edad):
    if edad < 12:
        return "Niño"
    elif edad >= 12 and edad < 18:
        return "Adolecente"
    elif edad >= 18 and edad < 60:
        return "Adulto"
    else:
        return "Adulto Mayo"
#Aplicar la funcion a la columna Edad
df_titanic2["Edad"] = df_titanic2["Edad"].apply(edad_categoria)
df_titanic2["Edad"].value_counts()
```
