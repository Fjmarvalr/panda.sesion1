### Creación de subconjuntos

```python
#copia del df_titanic0
df_titanic1 = df_titanic0.copy()
```

```python
df_titanic1.head(3)
```

```python
# extraer edad y sexo
df_age_sex = df_titanic1[["Age", "Sex"]]
```

```python
df_age_sex.head(3)
```

```python
# muestrame los pasajeros que son mayores de 30 años   
df_titanic1[df_titanic1["Age"] > 30]
```

```python
# dime cuantos pasajeros son mayores de 30 años y son mujeres
df_titanic1[(df_titanic1["Age"] > 30) & (df_titanic1["Sex"]=="female")]
```

```python
# dime cuantos pasajeros son mayores de 30 años y son mujeres
df_titanic1[(df_titanic1["Age"] > 30) & (df_titanic1["Sex"]=="female") & (df_titanic1["Survived"]==1)]
```

```python
df_female_30=df_titanic1[(df_titanic1["Age"] > 30) & (df_titanic1["Sex"]=="female")]
```

```python
#cuantas mujeres les faltan edad  
df_female_sinedad=df_titanic1[df_titanic1["Age"].isnull() & (df_titanic1["Sex"]=="female")]
```
